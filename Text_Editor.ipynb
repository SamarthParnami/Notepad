```python
from tkinter import *
from tkinter.filedialog import *
from configparser import ConfigParser
import re
labels=ConfigParser()
labels.read("literal.ini");

class Notepad:
  def __init__(self):
    self.file=None
    self.changedContent=False
    self.text

    self.main=Tk()
    self.elements()
    self.config()
    self.pack()
    self.main.mainloop()

  def elements(self):
    self.setupMenuBar()
    self.scrolly = Scrollbar(self.main)
    self.scrollx = Scrollbar(self.main, orient = HORIZONTAL)
    
    self.textWindow=Text(self.main,font="Consolas",undo=True,wrap=NONE,yscrollcommand = self.scrolly.set,
                            xscrollcommand = self.scrollx.set)
    



  def config(self):
    self.title="Untitled.txt"
    self.titleSuffix="  - Notepad"
    if(not self.file):
      self.main.title(self.title+self.titleSuffix)
    else:
      filename=re.search("[^////|\\\\]*\.[\w]*$",self.file).group()
      print("file",filename)
      self.main.title(filename+self.titleSuffix)
    self.scrolly.config(command = self.textWindow.yview)
    self.scrollx.config(command = self.textWindow.xview)



  def setupMenuBar(self):

    #adding menubar to window
    
    self.menuBar=Menu(self.main,tearoff=2)
    self.main.config(menu=self.menuBar)

    #First Menu Item
    self.menuItem1=Menu(self.menuBar)
    self.menuBar.add_cascade(label="File",menu=self.menuItem1)
    self.menuItem1.add_command(label = 'New', command = self.startFresh)
    self.menuItem1.add_command(label = 'New Window', command = self.newWindow)
    self.menuItem1.add_command(label = 'Open',command = self.openFile)
    self.menuItem1.add_command(label = 'Save',command = self.saveFile)
    self.menuItem1.add_command(label = 'Save As',command = self.saveFileAs)
    self.menuItem1.add_separator()
    self.menuItem1.add_command(label = 'Exit',command = self.exitNotepad)

    #Second Menu Item
    self.menuItem2=Menu(self.menuBar)
    self.menuBar.add_cascade(label="Edit",menu=self.menuItem2)
    self.menuItem2.add_command(label = 'Undo',command = self.undo)
    self.menuItem2.add_separator()
    self.menuItem2.add_command(label = 'Cut', command = self.cut)
    self.menuItem2.add_command(label = 'Copy', command = self.copy)    
    self.menuItem2.add_command(label = 'Paste',command = self.paste)
    self.menuItem2.add_command(label = 'Delete',command = self.delete)
    self.menuItem2.add_separator()
    self.menuItem2.add_command(label = 'Find',command = self.find)
    self.menuItem2.add_command(label = 'Replace',command = self.replace)
    self.menuItem2.add_command(label = 'Goto',command = self.goto)

    #Third Menu Item
    self.menuItem3=Menu(self.menuBar)
    self.menuBar.add_cascade(label="Format",menu=self.menuItem3)
    self.menuItem3.add_command(label = 'Font', command = self.fontSelector)
    self.menuItem3.add_command(label = 'Zoom', command = self.zoom)
  
  def pack(self):
    if not self.textWindow.see("end"):
      self.scrolly.pack(side = RIGHT, fill = Y)
    self.textWindow.pack(expand=True,fill=BOTH)
    if not self.textWindow.see("1.end"):
      self.scrollx.pack(side = BOTTOM, fill = X)
    
    
  def grid(self):
    pass
  def startFresh(self):
    print('New')
  
  def newWindow(self):
    print('New Window')
    Notepad()
  def openFile(self):
    print('open')
    filepath=askopenfilename(filetypes=[("Text Document","*.txt"),("All Files","*.*")])
    self.file=filepath
    print(self.file)
    
    file=open(filepath,'r')
    text=file.read()
    file.close()
    self.textWindow.delete("1.0","end")
    self.textWindow.insert("1.0",text)
    self.config()
  def saveFile(self):
    print('save') 
    file=open(self.file,"w")
    text=self.textWindow.get("1.0","end")
    print(text)
    file.write(text)
    file.close

  def saveFileAs(self):
    print('saveAs')
    file=asksaveasfilename(filetypes=[("Text Document","*.txt"),("All Files","*.*")],defaultextension=[("Text Document","*.txt"),("All Files","*.*")])
    if(file):
      self.file=file
      self.saveFile()
  def exitNotepad(self):
    print("Exit")

  def undo(self):
    print('Undo')
  
  def cut(self):
    print("cut")

  def copy(self):
    print("copy")

  def paste(self):
    print("paste")
  
  def delete(self):
    print("delete")
  
  def find(self):
    print("find")
  
  def replace(self):
    print("replace")

  def goto(self):
    print("goto")

  def fontSelector(self):
    print("font")

  def zoom(self):
    print("zoom")
notepad=Notepad()
```
